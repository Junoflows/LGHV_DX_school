```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

# opencv      ndarray  
#  2,  3
src1 = cv2.imread('./data/lena.jpg', cv2.IMREAD_GRAYSCALE)
src2 = np.zeros(shape = (512, 512), dtype =np.uint8) + 100

# dst = src1 + src2 #    
#    0 ~ 255  
# 256   256     
dst = cv2.add(src1, src2) # opencv    
# 256   255 
plt.imshow(dst, cmap = 'gray')
plt.axis('off')
plt.show()
```

```python
src = cv2.imread('./data/minMax.jpg', cv2.IMREAD_GRAYSCALE)
(min_val, max_val, _, _) =cv2.minMaxLoc(src)
print(min_val, max_val)

ratio = 255 / (max_val - min_val)
dst = np.round((src - min_val) * ratio).astype('uint8')
(min_dst, max_dst, _, _) = cv2.minMaxLoc(dst)
print(min_dst, max_dst)

# plt.imshow(src, cmap = 'gray')
plt.imshow(dst, cmap = 'gray')
plt.axis('off')
plt.show()
```
